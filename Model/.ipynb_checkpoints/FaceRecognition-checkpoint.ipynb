```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### 1. Process Data

```python
from sklearn.datasets import fetch_lfw_people

lfw_people = fetch_lfw_people(color=False, min_faces_per_person=20)

images = lfw_people.images # Tất cả các ảnh 
y = lfw_people.target # Nhãn mục tiêu
target_names = lfw_people.target_names # Danh sách tên của các đối tượng
```

```python
# Hiển thị một số hình ảnh mẫu từ bộ dữ liệu
fig, axes = plt.subplots(3, 5, figsize=(15, 8), subplot_kw={'xticks':(), 'yticks':()})
for i, ax in enumerate(axes.flat):
    ax.imshow(lfw_people.images[i], cmap='gray')
    ax.set_title(target_names[y[i]])

plt.show()

print(f"Tổng số cá nhân: {target_names.shape[0]}")
print(f"Tổng số ảnh: {images.shape[0]}")
```

```python
data_path = '../DataSet'
for person_id, person_name in enumerate(target_names):
    save_path = os.path.join(data_path, person_name)
    if not os.path.exists(save_path):
        os.makedirs(save_path)

    person_imgs = images[y == person_id]

    for idx, img in enumerate(person_imgs):
        img_path = os.path.join(save_path, 'img' + '_' + str(idx) + '.jpg')
        img_uint8 = (img*255).astype('uint8')
        cv2.imwrite(img_path, img_uint8)

print('LẤY DỮ LIỆU THÀNH CÔNG')
```

### 2. Preprocess Data 

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import OneHotEncoder
```

```python
X = np.array(images)
X = np.expand_dims(X, axis=-1)
```

```python
X.shape # Số lương ảnh, chiều cao ảnh, chiều rộng ảnh, kênh màu
```

```python
y = to_categorical(y).astype(int)
```

```python
y[0]
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

### 3. Build and Train Model CNN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D
from tensorflow.keras.layers import MaxPooling2D
from tensorflow.keras.layers import BatchNormalization
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
from keras.optimizers import Adam
import tensorflow as tf
```

```python
model = Sequential()

model.add(Conv2D(64, kernel_size=3, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2], 1)))
model.add(BatchNormalization())
model.add(Conv2D(64, kernel_size=3, activation='relu'))
model.add(BatchNormalization())
model.add(Conv2D(64, kernel_size=5, padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(128, kernel_size=3, activation='relu'))
model.add(BatchNormalization())
model.add(Conv2D(128, kernel_size=3, activation='relu'))
model.add(BatchNormalization())
model.add(Conv2D(128, kernel_size=5, padding='same', activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))

model.add(Conv2D(256, kernel_size=3, activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(128, activation='relu'))
model.add(BatchNormalization())
model.add(Dense(y.shape[1], activation='softmax'))

model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
early_stop = EarlyStopping(
    monitor='val_loss',   
    patience=10,          
    restore_best_weights=True 
)
```

```python
epochs = 14
batch_size = 32

history = model.fit(X_train, y_train, batch_size=batch_size,
           epochs=epochs,
           validation_data=(X_test, y_test),
           callbacks=[early_stop])
```

### 4. Evaluate Model

```python
# Đánh giá mô hình trên tập kiểm tra
loss, acc = model.evaluate(X_test, y_test)

# In ra kết quả
print(f'Loss: {loss}\nAccuracy: {acc * 100:.2f}%')
```

```python
model.save('model82.keras')
```

```python
# Vẽ đồ thị độ chính xác của mô hình
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()

# Vẽ đồ thị độ mất mát của mô hình
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
from sklearn.metrics import classification_report

print(classification_report(np.argmax(y_test, axis=1), np.argmax(model.predict(X_test),axis=1)))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_true = np.argmax(y_test, axis=1)
y_pred = np.argmax(model.predict(X_test), axis=1)

# Tính toán ma trận nhầm lẫn
conf_matrix = confusion_matrix(y_true, y_pred)

# Hiển thị ma trận nhầm lẫn với kích thước lớn hơn và điều chỉnh kiểu hiển thị
plt.figure(figsize=(30, 20))  # Điều chỉnh kích thước của biểu đồ
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=target_names, yticklabels=target_names,
            cbar=True, annot_kws={"size": 10})  # Thêm số lượng vào ô

# Thiết lập tiêu đề và nhãn trục
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.xticks(rotation=90)  
plt.yticks(rotation=0)   

plt.show()
```

### 5. Test In Real Time

```python
from tensorflow.keras.models import load_model

model_path = 'model84.keras'

# Load model
model = load_model(model_path)
```

```python

# Tải haar cascade
detector = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

# Mở webcam
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret or frame is None:
        print("Không thể lấy được khung hình từ webcam.")
        break

    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)  # Chuyển đổi sang ảnh xám
    faces = detector.detectMultiScale(gray_frame, scaleFactor=1.3, minNeighbors=5)
    
    # Dự đoán và vẽ hộp chữ nhật
    for x, y, w, h in faces:
        face_img = gray_frame[y:y+h, x:x+w]
        face_img = cv2.resize(face_img, (47, 62))  # Thay đổi kích thước

        face_img = face_img.astype('float32') / 255.0  # Chuẩn hóa ảnh
        face_img = np.expand_dims(face_img, axis=0)    # Thêm batch dimension
        face_img = np.expand_dims(face_img, axis=-1)   # Thêm kênh màu

         # Dự đoán
        predictions = model.predict(face_img)
        predicted_class = np.argmax(predictions)
        confidence = np.max(predictions)  # Lấy xác suất cao nhất

        # Xác định nhãn dự đoán
        if confidence * 100 < 60:
            predicted_label = "unknown"
        else:
            predicted_label = target_names[predicted_class]
        
        # Vẽ hộp chữ nhật và nhãn với xác suất
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)
        cv2.putText(frame, f"{predicted_label} ({confidence * 100:.2f}%)", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)
    
    
    # Hiển thị video
    cv2.imshow('Detected Face', frame)
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        print("Quá trình bị dừng lại bởi người dùng")
        break

cap.release()
cv2.destroyAllWindows()
```
